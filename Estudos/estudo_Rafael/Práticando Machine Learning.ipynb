```python
import pandas as pd
import numpy as np
%pylab inline
```

```python
data = pd.read_csv("california_housing_train.csv")
data.head()
```

```python
from sklearn.tree import DecisionTreeRegressor, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
X = data.drop("median_house_value", axis=1)
y = data['median_house_value']

Xtr, Xval, ytr, yval = train_test_split(X, y, test_size=0.5, random_state=0)

arvere = DecisionTreeRegressor(max_depth=3, random_state=0)

arvere.fit(Xtr, ytr)

p = arvere.predict(Xval)

np.sqrt(mean_squared_error(yval, p))
```

```python
pylab.figure(figsize=(60,40))
plot_tree(arvere, feature_names=Xtr.columns , filled = True);
```
